## Part 3) - Model Implementation

```python
#Implementation of the gaussian equation
def gaussian(x, mu, sig):
    return 1./(np.sqrt(2.*np.pi)*sig)*np.exp(-np.power((x - mu)/sig, 2.)/2)

#Implementation of a method to return values of the ouput as a numpy array
#this doesn't fly buddy .. im offended
def gaussian_vector(x,mu,sig):
    values = np.array([])
    values = np.append(values, gaussian(x, mu, sig))
    return np.array(values)
```

### Part V ) - Error between Model and Data

### Equation: Mean Absolute Error
$${\displaystyle \mathrm {MAE} ={\frac {\sum _{i=1}^{n}\left|y_{i}-x_{i}\right|}{n}}={\frac {\sum _{i=1}^{n}\left|e_{i}\right|}{n}}.}$$

### Model Predictions:
    - The model prediction(s) variable y is the model predictions.
### Dependent Variable(s):
    - The dependent variable is the MAE which is the output.

```python
def mean_absolute_error(features,response):
    size = len(features)
    sums = 0
    for i in range(0,size):
        sums += response[i]-features[i]
    sums = sums / 2
    return sums
```

```python
#getting the data(vector) which is the vector output for a given parameter set
firstp = mean_absolute_error(gaussian_vector(features,1,0.75),response)
secondp = mean_absolute_error(gaussian_vector(features,1,1.25),response)
thirdp = mean_absolute_error(gaussian_vector(features,5,1.25),response)
fourthp = mean_absolute_error(gaussian_vector(features,6,1.25),response)
```

```python
#Legend is the mean absolute error
plt.scatter(response, gaussian_vector(features,1,0.75), label="Firth Parameter %f" %firstp)
plt.scatter(response, gaussian_vector(features,1,1.25), label="Second Paramter %f" %secondp)
plt.scatter(response, gaussian_vector(features,5,1.25), label="Third Paramter %f" %thirdp)
plt.scatter(response, gaussian_vector(features,6,1.25), label="Fourth Parameter %f" %fourthp)
plt.xlabel("Response")
plt.ylabel("Gaussian Vector Output Values")
plt.legend()
plt.title("Response vs Gaussian Vector Values")
```

## MSE (Mean Squared Error)
### Equation: $${\displaystyle \operatorname {MSE} ={\frac {1}{n}}\sum _{i=1}^{n}(Y_{i}-{\hat {Y_{i}}})^{2}.}$$

### Model Predictions:
    - The model prediction variable(s) are "y hat" because it is the variable that is being used within the sample MSE when the predictor is being computed.

### Dependent Variables:
    - The dependent variable of this is "y" because "y" is the vector of the observed values of the variable being predicted 

### Independent Variables: 
    - The independent variable is the output from the equation which is MSE.

```python
#Implementation of MSE 
def mean_squared_error(true, predicted):
    squared_error = (predicted-true) **2
    sum_squared_error = np.sum(squared_error)
    loss = sum_squared_error/ true.size
    return loss
```

```python
#plot all on the same graph with all of the means on it 
newspaper_mean_error = mean_squared_error(sales,linear_regression(newspaper,b1,e1))
radio_mean_error = mean_squared_error(sales,linear_regression(radio,b1,e1))
tv_mean_error = mean_squared_error(sales,linear_regression(tv,b1,e1))
offset_mean_error = mean_squared_error(sales, linear_regression(offset,b1,e1))
```
