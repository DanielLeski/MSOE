### Imports

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from decision_boundaries import *
from sklearn.metrics import accuracy_score
```

### Experiment 1 - Decision Boundaries and Evaluation of Model Predictions with Metrics

#### a) Load the "setosa_data.csv" using the pandas read_csv() function

```python
setosa = pd.read_csv("setosa_data.csv")
```

#### b) Display the first few rows of the DataFrame using the head() function

```python
setosa.head()
```

#### c-d) Create a scatter plot of the data set points with the sepal length on the horizontal axis and the sepal width onthe verical axis. Plottting the specified color and the label the decision boundaries defined below

```python
color = setosa['label'] == "setosa"
plt.scatter(setosa['sepal_length (cm)'], setosa['sepal_width (cm)'], c=color, label='setosa')                                                                                           
sample = np.linspace(4,8,6)
db = 2 * sample - 8
db2 = 0.75 * sample - 0.9
db3 = -0.2 * sample + 12
plt.plot(sample, db, color="black")
plt.plot(sample, db2, color="red")
plt.plot(sample, db3, color="purple")
plt.title("Sepal Length vs Sepal Width")
plt.xlabel("Sepal Length")
plt.ylabel("Sepal Width")
plt.legend()
```

### f) Rewrite each decision boundary equation in a modified standard form
#### Equation: y = 2x-8 
   - -2x + y - 8

#### Equation: y = 0.75x - 0.9
   - -0.75 + y + 0.9

#### Equation : y = -0.2x + 12
   - -0.2x + y - 12 

#### G)  Import the linear_decision_boundary_classifier() function from the provided decision_boundaries.py file. Call the function with each pair of features and its associated decision boundary.

```python
eq1 = np.array([-2,1,8])
eq2 = np.array([-0.75, 1, 0.9])
eq3 = np.array([-0.2, 1, -12])

features = setosa[['sepal_length (cm)', 'sepal_width (cm)']].values
true_labels = setosa['label'].values

pred_label1 = linear_decision_boundary_classifier(eq1, features, true_labels, features)
accuracy_score1 = accuracy_score(true_labels, pred_label1)
print("Pred Labels 1: ", pred_label1)
print("Accuracy: ", accuracy_score1)

pred_label2 = linear_decision_boundary_classifier(eq2, features, true_labels, features)
accuracy_score2 = accuracy_score(true_labels, pred_label2)
print("Pred Labels 2 : ", pred_label2)
print("Accuracy : ", accuracy_score2)

pred_label3 = linear_decision_boundary_classifier(eq3, features, true_labels, features)
accuracy_score3 = accuracy_score(true_labels, pred_label3)
print("Pred Labels 3 : ", pred_label3)
print("Accuracy : ", accuracy_score3)
```
